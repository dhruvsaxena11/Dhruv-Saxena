```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

# tokenizer = AutoTokenizer.from_pretrained("deepset/roberta-base-squad2")
model = AutoModelForQuestionAnswering.from_pretrained("deepset/roberta-base-squad2")
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("question-answering", model="deepset/roberta-base-squad2")
```

```python
question = "What is the capital of France?"
context = "Paris is the capital of France."
result = pipe(question=question, context=context)
# Print the result
print(result["answer"])
```

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
prompt=PromptTemplate(
    input_variables=["question"],
    template="my question is that {question}?"
)
```

```python
import gradio
def ans(qn,cont):
  my=pipe(question=prompt.format(question=qn),context=cont)
  return my["answer"]
# an=ans("my name","my name is dhruv")
# print(an)
with gradio.Blocks() as dhruv:
    gradio.Markdown("""
    # AI Based Qna with your Text
    Add your context and question
    """)
    with gradio.Row():
        inp1 = gradio.Textbox(placeholder="Add Context",label="Context")
        inp2 = gradio.Textbox(placeholder="Add Question",label="Question")
        out=gradio.Textbox(label="Answer")
    btn1 = gradio.Button("Run")
    btn2= gradio.ClearButton([inp1,inp2])

    btn1.click(fn=ans, inputs=[inp2,inp1], outputs=out)

dhruv.launch()
```
